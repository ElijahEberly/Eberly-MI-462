### 2.1 Variables Explained

Below is a list of each variable in the dataset and its explanation.

* **Arena:** The name of the basket ball arena.
* **School:** The school that owns/plays at the arena.
* **City:** The city in which the arena is in.
* **State:**The state the arena is in
* **Conference:** The conference the school is in for basketball
* **Capacity:** the number of people that can fit in the arena
* **Opened:** When the arena opened
* **Wins:** # of wins the team had in 2017
* **Losses:** # of losses the team had in 2017
* **Conference Wins:** # of conference wins the team had in 2017
* **Conference Losses:** # of conference losses the team had in 2017
* **PPG_Scored:** average # of points scored in the season
* **PPG_Allowed:** average # of points allowed in the season
* **Simple Rating System:** a rating that takes into account average point differential and strength of schedule.
* **Strength of Schedule:** how hard the schedule is based on past years, and how the opposing teams are projected to do
* **NCAA Tourney:** # of times they have hosted an NCAA tournament
* **Home Games:** # off home games hosted at the arena
* **Total Attendance:** Total # of attendees throughout the season
* **Average Attendance:** average # of attendees per game
* **FTE:** Full time enrollment at the school
* **Attendence Rate:** average rate at which the arena is filled each game

```python
conf_wins = arenas.groupby("Conference", as_index = False).mean()
conf_wins = conf_wins[["Conference","Wins"]]
conf_wins = conf_wins.sort_values(by = "Wins", ascending = False)
arenas = arenas.rename(columns={'Attendence Rate':'Attendance Rate'})
```

```python
Conference = arenas.groupby(['Conference'], as_index = False).mean()
Conference_FTE_Win = Conference[['Conference', 'FTE', 'Wins']]
```

```python
corr_df = arenas.corr()
PPG_Allowed_Corr = corr_df[['PPG_Allowed']].loc['Strength of Schedule']

```

```python
conference = arenas.groupby('Conference', as_index = False).sum()
conference = conference[(conference['NCAA Tourney'] > 1)]
conference = conference[['Conference', 'NCAA Tourney']]
conference['percent'] = conference['NCAA Tourney'] / conference['NCAA Tourney'].sum()*100
conference = conference.sort_values(by = 'percent', ascending = False)
```

```python
Conference_FTE_Win_Power = Conference_FTE_Win[(Conference_FTE_Win['Conference'] == 'ACC') | (Conference_FTE_Win['Conference'] == 'Big Ten') |
                                              (Conference_FTE_Win['Conference'] == 'Big 12') | (Conference_FTE_Win['Conference'] == 'Pac-12') |
                                              (Conference_FTE_Win['Conference'] == 'SEC')]
```

```python
capacity_arenas = arenas.sort_values(by = "Capacity", ascending = False)
capacity_arenas = capacity_arenas.groupby("Conference").mean()
capacity_arenas = capacity_arenas.sort_values(by = "Capacity", ascending = False)
```

```python
grouped_arenas = arenas.groupby("Conference").sum()
count_arenas = arenas.groupby("Conference").count()
count_arenas
grouped_arenas = grouped_arenas.sort_values("Attendance Rate", ascending = False)
#grouped_arenas
count_arenas
grouped_arenas["Average Rate of Attendance"] = grouped_arenas["Attendance Rate"]/ count_arenas["School"]
grouped_arenas = grouped_arenas.sort_values("Average Rate of Attendance", ascending = False)
```

```python
plt.figure(figsize =(35,10))
plt.rcParams['font.size'] = '25'
plt.xticks(rotation = 90)

barlist = plt.bar(capacity_arenas.index, capacity_arenas["Capacity"])
plt.title("Average Arenas Capacity Per Conference")
plt.ylabel("Average Total Capacity")
plt.xlabel("Conference")
barlist[0].set_color('r')
barlist[1].set_color('r')
barlist[4].set_color('r')
barlist[6].set_color('r')
barlist[2].set_color('r')
```
